```python
import cv2 as cv
from time import sleep
cap = cv.VideoCapture("sincubre.mp4")
i=0
maskFace = cv.CascadeClassifier('Dataset1/classifier/cascade.xml')
while cap.isOpened():
    ret, im = cap.read()
    
    if ret==False:
        break
        
    height, width, _ = im.shape

    # Calcula las coordenadas del rectángulo para que esté en el centro
    x1 = int(width / 2) - 200  # 150 es la mitad del ancho del rectángulo
    y1 = int(height / 2) - 250  # 150 es la mitad del alto del rectángulo
    x2 = int(width / 2) + 200
    y2 = int(height / 2) + 200
    
    print(x1,x2,y1,y2)
    cuadro = im[y1:y2, x1:x2]
    frameG = cv.cvtColor(cuadro, cv.COLOR_BGR2GRAY)
    cara = maskFace.detectMultiScale(frameG,
    scaleFactor = 9, #Tambien se maneja con scale Factor: 10 y minNeighbors: 200
    minNeighbors = 335,
    minSize = (200,200))

    for (x,y,w,h) in cara:
        cv.rectangle(cuadro, (x,y),(x+w,y+h),(0,255,0),2)
        cv.putText(cuadro,'Con cubrebocas',(x,y-10),2,0.7,(0,255,0),2,cv.LINE_AA)
    
    cv.imshow('Video', im)
    cv.imshow('Cuadro', cuadro)
    k=cv.waitKey(1)
    if k == 27:
        break
    #sleep(1/30)
cap.release()
cv.destroyAllWindows() 
```
